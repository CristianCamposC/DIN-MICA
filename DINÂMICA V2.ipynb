#### 3.1. AÇÃO DO VENTO EM TORRES DE COMUNICAÇÃO


Ventos são movimentos livres das massas de ar provocados pelas diferenças de pressão e temperatura das camadas da atmosfera. Trata-se de um processo de redistribuição da energia solar através dos movimentos dos gases que compõem as camadas da atmosfera (LOPES, 2003). JERVELL (2008) complementa que, este processe ocorre, em primeiro lugar, devido a radiação solar recebida na Terra ser maior nas zonas equatoriais do que nas regiões polares e, em segundo, pelo movimento de rotação da Terra que ocasiona variações sazonais de distribuição da energia solar incidente. 

Os ventos são considerados grandezas vetoriais, portanto, possuem um componente escalar (quantificação da intensidade) e um componente vetorial (indicador direcional). A velocidade dos ventos pode ser expressa em quilômetros por hora (km/h), milhas por hora (mph), nós/milhas náuticas por hora (kt) ou metros por segundo (m/s) no Sistema Internacional de Medidas (SI). Dependendo da aplicação, utiliza-se uma destas unidades. Em pesquisas científicas e em projetos de engenharia, por exemplo, para a análise dos efeitos dinâmicos do vento em estruturas, utiliza-se metros por segundo como unidade principal de velocidade.

CARRIL (2000) quando estuda os efeitos dinâmicos em torres treliçadas para telecomunicações afirma que os possíveis esforços que podem estar atuando nesse tipo de estrutura são os esforços pseudo estáticos correspondentes à parte média do vento e os esforços dinâmicos decorrentes da parte flutuante do vento. O cálculo do efeito dinâmico do vento sobre estruturas esbeltas, conforme CARRIL (2000), é composto de três estágios básicos: 

i) a descrição do vento; <br/>
ii) a descrição das propriedades físicas e aerodinâmicas da estrutura; <br/>
iii) a combinação desses fatores na determinação da resposta da estrutura. <br/>

A NBR 6123 (ABNT, 1988) incorpora esses conceitos e destaca que edificações com período fundamental superior a 1s, frequências até 1 Hz, podem apresentar importante resposta flutuante na direção do vento médio. Neste sentido, a norma apresenta três modelos de cálculo para a ação do vento nas estruturas, assim denominados: 
i) Forças estáticas devidas ao vento ou Modelo Estático; <br/>
ii) Modelo dinâmico simplificado; <br/>
iii) Modelo dinâmico discreto. <br/>

Neste trabalho, será adotado o método da NBR 6123 (ABNT, 1988) para o cálculo das forças estáticas devidas ao vento. Para a análise dinâmica da torre, o cálculo da densidade espectral de potência baseou-se no método proposto por Davenport, enquanto que para as componentes flutuantes foi utilizado a superposição de ondas harmônicas proposta por Shinozuka e Jan (1972).

### 5. ANÁLISE ESTRUTURAL

#### 5.1. MATRIZES DE RIGIDEZ, MASSA E AMORTECIMENTO

O primeiro passo para a determinação do problema é a leitura dos dados coordenadas nós, conetividade e as propriedades de cada barra, como seção, comprimento e Módulo de _Young_. Para isto, utilizou-se os dados salvos em planilha Excel existentes e já utilizados para a criação dos modelos em _AutoCAD e ANSYS_ anteriormente ilustrados.

```python
nos = pd.read_excel (r'data\nos.xlsx')  # Importa a planilha com as coordenadas dos nós
dados = nos.values                      # Transforma os dados da planilha em uma matriz

x = dados[:,1]
y = dados[:,2]

cone = pd.read_excel (r'data\elementos.xlsx')
dados2 = cone.values

nele = len(dados2)  # Número de elementos (barras)
nnos = len(dados)   # Número de Nós

ni = dados2[:,1]    # Nós iniciais
nf = dados2[:,2]    # Nós finais
A  = dados2[:,4]    # Área de seção em m²
E  = dados2[:,3]    # Módulo de Elasticidade em N/m²
D  = dados2[:,5]    # Diâmetro das barras em m
xb = dados2[:,6]
yb = dados2[:,7]
EA = A*E            # Rigidez Axial

L = np.zeros(nele)  # Vetor que armazenará os comprimentos

Cx = np.zeros(nele) # Vetores que armazenarão os cossenos diretores
Cy = np.zeros(nele)
```

A partir de então, determinou-se as matrizes de rigidez e massa consistente da estrutura, necessárias para a análise do problema.

```python
KG  =  np.zeros((2*nnos,2*nnos))
MG  =  np.zeros((2*nnos,2*nnos))
for k in range (nele):
    
    I = ni[k] 
    F = nf[k]
   
    L = np.sqrt((x[int(I-1)]-x[int(F-1)])**2 + (y[int(I-1)]-y[int(F-1)])**2)

    Cx = (x[int(I-1)]-x[int(F-1)])/L
    Cy = (y[int(I-1)]-y[int(F-1)])/L
    
    k11=(Cx**2)*(EA[k]/L)
    k22=(Cy**2)*(EA[k]/L)
    k12=(Cx*Cy)*(EA[k]/L)

    a=abs(int((2*I)-2))
    b=abs(int((2*I)-1))
    c=abs(int((2*F)-2))
    d=abs(int((2*F)-1))
    
    KG[a,a]=KG[a,a]+k11;   KG[a,c]=KG[a,c]-k11
    KG[b,b]=KG[b,b]+k22;   KG[b,d]=KG[b,d]-k22
    KG[a,b]=KG[a,b]+k12;   KG[a,d]=KG[a,d]-k12
    KG[b,a]=KG[b,a]+k12;   KG[b,c]=KG[b,c]-k12
    
    KG[c,a]=KG[c,a]-k11;   KG[c,c]=KG[c,c]+k11
    KG[d,b]=KG[d,b]-k22;   KG[d,d]=KG[d,d]+k22
    KG[d,a]=KG[d,a]-k12;   KG[c,d]=KG[c,d]+k12
    KG[c,b]=KG[c,b]-k12;   KG[d,c]=KG[d,c]+k12
    
    C=(7850*A[k]*L)/6
    
    MG[a,a]=MG[a,a]+(2*C);   MG[a,c]=MG[a,c]+C
    MG[b,b]=MG[b,b]+(2*C);   MG[b,d]=MG[b,d]+C
    
    MG[c,a]=MG[c,a]+C;   MG[c,c]=MG[c,c]+(2*C)
    MG[d,b]=MG[d,b]+C;   MG[d,d]=MG[d,d]+(2*C)

    
#Eliminando das matrizes os graus de liberdade restringidos

KG = KG[4:,4:]
MG = MG[4:,4:]
```

#### 5.2. FREQUÊNCIAS NATURAIS E MODOS DE VIBRAÇÃO

De posse das matrizes de rigidez e massa, determina-se a frequência fundamental da estrutura por meio da formulação em _Python_ para compará-la com os resultados obtidos através da análise no software _ANSYS_.

```python
import scipy.linalg as sc
ava,ave   =  sc.eig(KG, MG)

avai  = ava.argsort()
ava  = ava[avai]
ave = ave[:,avai]


wk  = np.sqrt(np.real(ava)) 
fk  = wk/2/np.pi

print(fk[0:9])
```

Deste modo, verifica-se que as frequências entre os dois métodos de análise são idênticas.

Para a análise dinâmica estrutural, torna-se necessário determinar uma relação de amortecimento para a estrutura. Neste trabalho, considera-se que o amortecimento estrutural esteja de acordo com a metodologia de Rayleigh, ou seja, que o amortecimento seja proporcional à rigidez e à massa da estrutura, conforme a seguir.

```python
L = np.zeros(nele)  # Vetor que armazenará os comprimentos

for q in range(nele):
    ni1 = ni[q] 
    nf1 = nf[q]
    L[q] = np.sqrt((x[int(ni1-1)]-x[int(nf1-1)])**2 + (y[int(ni1-1)]-y[int(nf1-1)])**2)
```

#### 5.3. VELOCIDADE MÉDIA DO VENTO

Para os parâmetros de projeto adotados considerou-se:

```python
# CÁLCULO DA VELOCIDADE MÉDIA DO VENTO

def Vmedia(z):
    b = .71               # Constante de Karman
    p = .23               # Rugosidade Categoria IV
    Fr = .69              # Altura de referência 
    V0 = 45             # Velocidade de referência (Classe C)
    S1 = 1; S3 = 1
    S2 = b*Fr*(z/10)**p
    Vm = S1*S2*S3*V0
    return Vm

Vmedia(10)
```

#### 5.4.1. ESPECTRO DE DAVENPORT

Para aplicar a superposição de ondas, deve-se definir primeiro o espectro de velocidade do vento. Segundo TROIAN (2018), o espectro de velocidade se refere aos dois lados do espectro, real e imaginário. O espectro de Davenport é definido através da seguinte equação

<img src="data/Davenport.png" alt="Estrutura" width="200px"/>


Sendo que esta expressão relafiona a frequência _f_ correspondente a um turbilhão, a energia _S_ contida em uma frequência específica, a velocidade de fricção e uma frequência _X_ adimensional com valores pré-estabelecidos. 

A literatura diverge a respeito da faixa de frequências que deve ser utilizada para considerar a modelagem do vento atuando em uma estrutura. Devido a isto, determina-se a faixa de limites compreendida entre _0,01Hz_ e _3,0Hz_, que é uma faixa utilizada em uma ampla gama de trabalhos já existentes. Como exemplo, DEODATIS (1996) utilizou _4,0Hz_, GATTULLI _et al_ (2007) utilizaram _2,0Hz_, MIGUEL _et al_ (2012) utilizaram _3,0Hz_, e BALLABEN _et al_ (2014) utilizaram _2,5Hz_.

Neste trabalho, por possuir fins acadêmicos, a faixa de frequências adotada foi de 0,0001 a 3 Hz, para que permita uma visualização completa do espectro de Davenport.

Além disto, o espectro de Davenport é um espectro independente da altura, deste modo, assume-se que a energia de flutuação é homogênea ao longo de toda a altura da estrutura analisada. Esta simplificação não está totalmente coerente com a realidade, uma vez que, segundo TROIAN (2018), a energia de flutuação diminui com a altura. Entretanto, diferentes autores já propuseram modifiações no espectro de Davenport para melhorar o modelo, direcionando-o a análises em diferentes terrenos ousimplesmente fazer ajustes em hipóteses assumidas para este espectro. De uma forma geral, os epectros de Kaimal, Harris e von Kárman estão entre os mais utilizados na atualidade, podendo ser facilmente encontrados na literatura especialidada. A Figura a seguir, retirada do trabalho de TROIAN (2018), ilustra a comparação entre os valores obtidos com diferentes espectros para a mesma análise. 

<img src="data/CompEspec.png" alt="Estrutura" width="500px"/>


```python
# CÁLCULO DA VELOCIDADE FLUTUANTE DO VENTO

# ESPECTRO DE DAVENPORT

f1 = 0.0001            # Frequência inicial
f2 = 3                 # Frequência final
nn = 5000              # Número de pontos

f = np.linspace(f1,f2,nn)           
z0 = 1                 # Comprimento de Rugosidade (Categoria 4)
V10 = Vmedia(10)       # Velocidade de Referência
n = f*1200/V10               
u = 0.4*V10/np.log(10/z0)
Sw = 4*n**2*u**2/(f*(1+n**2)**(4/3))

plt.plot(f,Sw)
plt.xlim((f1, f2))     # set the xlim to left, right
plt.xscale('log')
plt.title('Espectro de Davenport')
plt.xlabel('f (Hz)')
plt.ylabel('Sw')
```

#### 5.4.2. SHINOZUKA E JAN

Conforme SHINOZUKA e JAN (1972), o espectro de potências e velocidades, que neste caso é o espectro de Davenport, pode ser dividido em parcelas a partir das quais são gerados resultados em função das frequências e energias do escoamento. A equação utiliada para a análise deste processo é a equação a seguir

<img src="data/velocidade(S).png" alt="Estrutura" width="200px"/>


Tal expressão depende do número de frequências ao qual o espectro está dividido, o intervalo ao qual o espectro de velocidades foi discretizado, a frequência de cada função gerada, a discretização do tempo e a amplitude do espectro de potência de cada velocidade na sua respectiva frequência. O ângulo referente ao último termo dessa equação diz respeito a um valor aleatório para cada simulação, gerado em um intervalo definido.


```python
# SHINOZUKA E JAN

phi = np.zeros((nn))          # Criando vetor que armazenará ângulos de fase aleatórios

for p in range(nn):
    phi[p] = random.random()*2*np.pi

df = (f2-f1)/nn                 # Discretização do domínio da frequência
dt = (600-0)/500
t = np.linspace(0,600,500)     # Discretização do domínio do tempo

delta_V = np.zeros((len(f)))    # Vetor que armazenará as velocidades
Sinal = np.zeros((len(t)))      # Vetor que armazenará a parcela flutuante da velocidade


for tt in range(len(t)):
    for ff in range(len(f)):
        delta_V[ff] =np.sqrt(2*Sw[ff]*df)*np.cos(2*np.pi*f[ff]*t[tt]+phi[ff])
    Sinal[tt]=np.sum(delta_V)
    

plt.plot(t,Sinal)
plt.title('Componente Flutuante da Velocidade do Vento')
plt.xlabel('t (s)')
plt.ylabel('Velocidade Flutuante (m/s)')
plt.grid(True)
plt.xlim(0, 600)
```

##### 5.4.3. CORRELAÇÃO ESPACIAL DO CAMPO DE VELOCIDADES

O método de correlação propõe uma maneira de gerar, através de sinais independentes, sinais correlacionados. Os pontos onde serão gerados os sinais de velocidade devem ser suficientemente espaçados para que a consideração de independência dos valores seja considerada válida. A figura a seguir ilustra  a idealização da malha para a determinação dos pontos onde haverá a análise direta dos valores.

<img src="data/Correlacao.png" alt="Estrutura" width="600px"/>


A análise com os pontos de correlação não necessariamente coloca pontos exatamente sobre a estrutura que está sendo analisada, que é o que estamos buscando. Deste modo, deve-se realizar uma interpolação entre os valores obtidos nos nós da malha para a obtenção das velocidades nos nós que compõe a estrutura. A equação a seguir ilustra esta análise para o caso bidimensional (o método foi desenvolvido por RIEIRA e AMBROSINI, em 1992).

<img src="data/Interpolação2D.png" alt="Estrutura" width="400px"/>


Deste modo, é possível determinar os valores da ação flutuante do vento incidindo sobre a estrutura, nos pontos especificados pela interpolação utilizada.

```python
Z = 150                         # Altura total da torre [m]

# Comprimentos de Correlação segundo Riera e Ambrosini (1992)

a = 1.6*Z + 22.1                # Vertical [m]
b = .93*Z + 29.3                # Horizontal [m]
xt = x + (a/2- x[1]/2)          # Ajustando as coordenadas do nó em relação à malha de correlação

# Cálculo da velocidade (Média + Flutuante) nos pontos da malha

V1 = 0
V2 = 0
V3 = Vmedia(b) + Sinal      
V4 = Vmedia(b) + Sinal 

# Cálculo da velocidade nos nós da torre:

V = np.zeros((nnos,len(Sinal)))  # Vetor que armazenará as velocidades em cada nó

for i in range(nnos):
    for j in range(len(Sinal)):
        V[i,j] = V1 + (V2-V1)*xt[i]/a + (V3[j]-V1)*y[i]/b + (V4[j]-V3[j]-V2-V1)*xt[i]*y[i]/(a*b)

# Exemplo de velocidades no nó 99:

plt.plot(t,V[100])
plt.grid(True)
plt.xlim(0, 600)
plt.xlabel('t (s)')
plt.ylabel('Velocidade (m/s)')
```

### TRANSFORMANDO A VELOCIDADE EM FORÇA  
Primeiramente, calculou-se a força no centro de cada uma das 200 barras, de acordo com a NBR 6123 (barras prismáticas de seção circular). Depois disso, a força de cada barra foi dividida por dois e transferida aos nós inicial e final. Com isso, obteve-se um vetor de registro de forças que atuam em cada nó, pelo tempo. 

Agora, calculando a velocidade em cada metade de barra:

```python
# Velocidade que atua no centro de cada barra

Vb = np.zeros((nele,len(Sinal)))
for i in range(nele):
    for j in range(len(Sinal)):
        Vb[i,j] = V1 + (V2-V1)*xb[i]/a + (V3[j]-V1)*yb[i]/b + (V4[j]-V3[j]-V2-V1)*xb[i]*yb[i]/(a*b)
```

```python
# Pressão dinâmica

q = 0.613*V**2      # Pressão Dinâmica (nos nós)
qb = 0.613*Vb**2    # Pressão Dinâmica (no centro das barras)

# Como Re < 4,2 x 10^5 para todas barras:

Ca = 1.2

# Fator de correção K

Ke = np.zeros(nele)
LD = L/D

ld = [2, 5, 10, 20, 40, 50, 100, 10000]            # NBR 6123
k = [.58, .62, .68, .74, .82, .87, .98, 1.0]       # NBR 6123

for d in range(nele):
    Ke[d] = np.interp(LD[d],ld,k)

# Força de Arrasto:

Fab = np.zeros((nele,500))

for i in range(nele):
    for tt in range(500):
        Fab[i,tt] = Ca*Ke[i]*L[i]*D[i]*qb[i,tt]

# Exemplo: Força atuando no centro da barra 200:

plt.plot(t, Fab[199])
plt.grid(True)
plt.xlim(0, 600)
plt.xlabel('t (s)')
plt.ylabel('F (N)')

np.max(qb[199,])
```

```python
# Força de arrasto em cada um dos nós

Fa = np.zeros((nnos,500))

for i in range(nele):
    Fa[int((dados2[i,1]-1)),] =  Fa[int((dados2[i,1]-1)),] + Fab[i,]/2
    Fa[int((dados2[i,2]-1)),] =  Fa[int((dados2[i,2]-1)),] + Fab[i,]/2

# Exemplo de força atuando no nó 100:

plt.plot(t, Fab[99])
plt.grid(True)
plt.xlim(0, 600)
plt.xlabel('t (s)')
plt.ylabel('F (N)')

VetF = np.zeros((nnos*2,500))

for s in range(nnos):
    VetF[((2*s)+1),] = Fa[s,]
```

Conhecendo o vetor de forças, pode-se resolver a equação do movimento. Para isso, optou-se pela Integração Numérica através do Método de Newmark.

```python
force_dof = np.zeros((nnos*2-4),int)
for i in range((nnos*2-4)):
    if i%2 == 0:
        force_dof[i] = i+1
    else:
        force_dof[i] = -999



Fa = Fa[2:]
```

```python
def Newmark_coefficients(dt):
    alpha=0.25
    beta=0.5

    a0=1/(alpha*(dt**2))
    a1=beta/(alpha*dt)
    a2=1/(alpha*dt)
    a3=(1/(2*alpha))-1
    a4=(beta/alpha)-1
    a5=(dt/2)*((beta/alpha)-2)
    a6=dt*(1-beta)
    a7=beta*dt

    return a0,a1,a2,a3,a4,a5,a6,a7


def Newmark_initialize(ndof,a0,a1,M,C,K,NT,DI,VI):

    nlength=len(np.atleast_1d(M)) 

    KH=np.zeros((ndof,ndof),float)

    KH=K+a0*M+a1*C

    if(nlength==1):

        U=np.zeros(NT,float)
        Ud=np.zeros(NT,float)
        Udd=np.zeros(NT,float)

        U[0]=DI
        Ud[0]=VI
            
    else:

        U=np.zeros((ndof,NT),float)
        Ud=np.zeros((ndof,NT),float)
        Udd=np.zeros((ndof,NT),float)

        U[:,0]=DI
        Ud[:,0]=VI

    return U,Ud,Udd,KH


def Newmark_force(M,C,K,FFI,force_dof,VI,DI,dt,NT,ndof):
    """
    input

    M = mass matrix
    C = damping matrix
    K = stiffness matrix

    FFI = interpolated force matrix
    force_dof = connects forces with dofs

    VI = initial velocity
    DI = initial displacement

      dt = time step
      NT = number of time points
    ndof = number of degrees of freedom

    output

      U = displacement
     Ud = velocity
    Udd = acceleration

    """

    a0,a1,a2,a3,a4,a5,a6,a7=Newmark_coefficients(dt)

    U,Ud,Udd,KH=Newmark_initialize(ndof,a0,a1,M,C,K,NT,DI,VI)


    for i in range (1,NT):

        V1=(a1*U[:,i-1]+a4*Ud[:,i-1]+a5*Udd[:,i-1])
        V2=(a0*U[:,i-1]+a2*Ud[:,i-1]+a3*Udd[:,i-1])

        CV=np.dot(C,V1)
        MA=np.dot(M,V2)


#  apply forces

        F=np.zeros(ndof,float)

        for j in range (0,ndof):

            j_index=force_dof[j]

            if(j_index!=-999):

                F[j]=FFI[i,j_index]


        FH=F+MA+CV

#  solve for displacements

        if(ndof>1):
            Un = np.linalg.solve(KH, FH)
        else:
            Un=FH/KH

        Uddn=a0*(Un-U[:,i-1])-a2*Ud[:,i-1]-a3*Udd[:,i-1]
        Udn=Ud[:,i-1]+a6*Udd[:,i-1]+a7*Uddn


        U[:,i]=Un
        Ud[:,i]=Udn
        Udd[:,i]=Uddn

    return U.T,Ud.T,Udd.T
```
